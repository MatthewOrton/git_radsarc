```python
import numpy as np
import pandas as pd
pd.options.display.max_rows = 20
pd.options.display.max_columns = 1000
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# load data
df = pd.read_csv('/Users/morton/Dicom Files/RADSARC_R/XNAT/extractions/extractions__20220910_1006_allRegions_incomplete/radiomicFeatures/radiomicFeatures.csv')
df.drop(list(df.filter(regex = 'source')), axis = 1, inplace = True)
df.drop(list(df.filter(regex = 'diagnostic')), axis = 1, inplace = True)
```

```python
# get strings for each region - expecting 'lesion', 'low_enhancing', 'mid_enhancing', high_enhancing', 'lesion_calcificationDeleted'
regions = list(set([x.split('_original_')[0] for x in df.columns]))
regions = [x for x in regions if 'StudyPatientName' not in x and 'sarcomaFeature' not in x]
```

```python
# take logs of some features
logFeatures = ['shape_MeshVolume', 'shape_SurfaceArea', 'shape_SurfaceVolumeRatio', 'shape_VoxelVolume', 
               'firstorder_Energy', 'firstorder_TotalEnergy', 'firstorder_Variance']
for feat in df.columns:
    if any([y in feat for y in logFeatures]):
        df.loc[:,feat + '_log10'] = np.log10(df[feat])
        df.drop(feat, axis=1, inplace=True)
```

```python
# get strings for radiomic features
featureNames = [x.replace(regions[0]+'_','') for x in df.filter(regex = regions[0]).columns]
```

```python
# package data with one column per feature, with each region row-stacked
dfFeat = pd.DataFrame()

regCol = []
for reg in regions:
    regCol += [reg]*df.shape[0]
dfFeat['region'] = regCol

for fName in featureNames:
    thisFeature = []
    for reg in regions:
        thisFeature += list(df.filter(regex=reg)[reg+'_'+fName])
    dfFeat[fName] = thisFeature
```

```python
f,a = plt.subplots(3,5, figsize=(20,12))
a = a.ravel()
for n, col in enumerate(dfFeat.filter(regex = 'shape', axis = 1).columns):
    sns.kdeplot(data=dfFeat, x=col, hue='region', ax = a[n], legend=(n==0))
```

```python
f,a = plt.subplots(3,6, figsize=(20,12))
a = a.ravel()
for n, col in enumerate(dfFeat.filter(regex = 'firstorder', axis = 1).columns):
    sns.kdeplot(data=dfFeat, x=col, hue='region', ax = a[n], legend=(n==0))
```
